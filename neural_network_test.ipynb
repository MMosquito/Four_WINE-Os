```python
import pandas as pd
import sklearn as skl
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
wine_df = pd.read_csv("preprocessed_wine_data.csv", sep=",")
wine_df.head(20)
```

```python
wine_df['quality'].value_counts()
```

```python
df2 = wine_df.sort_values('quality', ascending=False)
df2.head(20)
```

```python
check_decimals = lambda x: str(x).count('.') <= 1

remove_extra_decimals = wine_df['alcohol'].apply(check_decimals)

wine_df = wine_df[remove_extra_decimals]
```

```python
X = wine_df.drop("quality", axis=1).values
X
```

```python
y = wine_df["quality"].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_scaler = skl.preprocessing.StandardScaler()
X_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_train
```

```python
nn = tf.keras.models.Sequential()
number_input_features = len(X_train[0])

nn.add(tf.keras.layers.Dense(units=50, activation="relu", input_dim=number_input_features))
nn.add(tf.keras.layers.Dense(units=22, activation="relu"))
nn.add(tf.keras.layers.Dense(units=1, activation="relu"))
nn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
