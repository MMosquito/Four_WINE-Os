```python
import pandas as pd
```

```python
wine_df = pd.read_csv("wine.csv", sep=";")
wine_df.head(20)
```

```python
wine_df.dtypes
```

```python
wine_df['alcohol'].value_counts()
```

```python
alc = wine_df["alcohol"].tolist()
alc
```

```python
check_decimals = lambda x: str(x).count('.') <= 1

remove_extra_decimals = wine_df['alcohol'].apply(check_decimals)

wine_df = wine_df[remove_extra_decimals]
```

```python
wine_df['alcohol'] = wine_df['alcohol'].astype(float)
```

```python
df2 = wine_df.sort_values('density', ascending=False)
df2.head(50)
```

```python
check_density = lambda x: float(x) < 2

density_repair = wine_df['density'].apply(check_density)

wine_df = wine_df[density_repair]
```

```python
df5 = wine_df.sort_values('alcohol', ascending=False)
df5.head(50)
```

```python
wine_df.to_csv('preprocessed_wine_data.csv', index=False)
```
