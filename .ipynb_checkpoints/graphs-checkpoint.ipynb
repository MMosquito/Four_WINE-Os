```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Wine_RegionColor.csv")
df.head()
```

```python
df["quality"].value_counts().plot(kind='bar', xlabel="Quality Rating", ylabel="Number of Wines")
```

```python
quality_avgs_df = df.groupby(by=["quality"], as_index=False).mean()
quality_avgs_df
```

```python
quality_avgs_df.plot.bar(x="quality", y="residual_sugar", xlabel="Quality Rating", ylabel="Residual Sugar")
```
